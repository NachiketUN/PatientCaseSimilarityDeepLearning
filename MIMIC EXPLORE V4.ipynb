```python
from gensim.models import Word2Vec
import numpy as np
embedding = Word2Vec.load('model-mimic.bin')

def gen_vector(record):
    vector = np.zeros(50)
    record = record.split(' ')
    for word in record:
        vector += embedding[word]
    return vector
    
```
